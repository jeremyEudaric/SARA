```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_overview_file= '/hkfs/home/haicore/hgf_dlr/hgf_uyx7543/Results_flood_hazard.csv'
# read in overview file
df_flood_hazard= pd.read_csv(path_overview_file, encoding='utf-8')
df_flood_hazard.head()
```

```python
# Define a function to 
def intervalls(percentage):
    if percentage == 0:
        return 0
    
    if 0 <= percentage <= 25:
        return 1
    elif 25 < percentage <= 50:
        return 2
    elif 50< percentage <= 75:
        return 3
    elif 75< percentage <= 100:
        return 4
    else:
        return None  # Handle any other cases if needed
```

```python
path_overview_file= '/hkfs/home/haicore/hgf_dlr/hgf_uyx7543/Buildings_exposure.csv'
# read in overview file
df_buildings= pd.read_csv(path_overview_file, encoding='utf-8')
df_buildings.head()
```

```python
df_buildings["Buildings Exposure"] = df_buildings["Buildings exposure"].apply(intervalls)
```

```python
df_buildings
```

```python
df_buildings = df_buildings.drop(df_buildings.index[-1])
```

```python
path_overview_file= '/hkfs/home/haicore/hgf_dlr/hgf_uyx7543/Results_population_exposure.csv'
# read in overview fils
df_pop= pd.read_csv(path_overview_file, encoding='utf-8')
df_pop
```

```python
df_pop = df_pop.drop(df_pop.index[-2:])
```

```python
df_pop
```

```python
df_pop["population Exposure"] = df_pop["population exposure"].apply(intervalls)
```

```python
new_dataframe = pd.DataFrame({
    "Buildings Exposure": df_buildings["Buildings Exposure"],
    "population Exposure":df_pop["population Exposure"]
    
}).set_index(df_buildings["Municipality"])

new_dataframe["Exposure Index"] = new_dataframe["Buildings Exposure"]+new_dataframe["population Exposure"]
```

```python
new_dataframe
#new_dataframe.to_csv('EDAI(Sentinel-1).csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 5))

# Create the heatmap with specified vmin and vmax
sns.heatmap(new_dataframe[['Exposure Index']].T.astype(int), 
            cmap="YlOrRd", 
            linewidths=0, 
            annot=False, 
            vmin=0, 
            vmax=8, 
            cbar_kws={"ticks": [0,1, 2,3,4,5,6,7,8]})  # Set color bar ticks

# Improve the edge between the municipality names
plt.yticks(rotation=0, fontsize=15)  # Rotate the y-axis labels and increase font size
plt.xticks(fontsize=15)  # Increase font size of x-axis labels
plt.tight_layout()  # Adjust the layout to prevent clipping of labels
plt.savefig("Heat_map_EI.png")
```

```python
# Assuming df_flood_hazard is your DataFrame
# Step 1: Remove municipalities with a flood hazard value of 0%
df_flood_hazard = df_flood_hazard[df_flood_hazard['Flood_hazard'] > 0]

#Step 2: Sort the DataFrame by Municipality names in alphabetical order
#df_flood_hazard = df_flood_hazard.sort_values(by='Municipality')

fig, ax = plt.subplots(figsize=(50, 50))

# Calculate the height of each bar
bar_height = 0.5

# Calculate the y positions of the bars
y = np.arange(len(df_flood_hazard['Flood_hazard']))

# Create the bars with the specified height
bars = ax.barh(y, df_flood_hazard['Flood_hazard'], height=bar_height, align='center')

# Set the y-axis tick labels to the municipality names
ax.set_yticks(y)
ax.set_yticklabels(df_flood_hazard['Municipality'], fontsize=110)
ax.set_xlabel('Percentage of flood hazard', fontsize=110)

# Add space between the bars
ax.set_ylim(-bar_height, len(df_flood_hazard['Flood_hazard']) - bar_height)

# Adjust the figure layout to ensure all municipality names are fully visible
ax.spines[['right', 'top']].set_visible(False)

# Add the percentage values on the bars in black, excluding 0.0%
for bar in bars:
    value = bar.get_width()
    ax.text(value, bar.get_y() + bar.get_height() / 2, f'{value:.1f}%', ha='left', va='center', color='black', fontsize=100)

plt.tight_layout()
plt.savefig("Flood_hazard.png")
```

```python
# Step 1: Remove municipalities with 0% population exposure
df_pop = df_pop[df_pop["population exposure"] > 0]

# Step 2: Sort the DataFrame by municipality names alphabetically
df_pop = df_pop.sort_values(by="Municipality")

# Step 3: Sort by population exposure for plotting
# This step is not necessary if you want to keep the alphabetical order
# df_pop = df_pop.sort_values(by="population exposure", ascending=True)

fig, ax = plt.subplots(figsize=(50, 50))

# Calculate the height of each bar
bar_height = 0.5

# Calculate the y positions of the bars
y = np.arange(len(df_pop["population exposure"]))

# Create the bars with the specified height and color
bars = ax.barh(y, df_pop["population exposure"], height=bar_height, align='center')

# Set the y-axis tick labels to the municipality names
ax.set_yticks(y)
ax.set_yticklabels(df_pop['Municipality'], fontsize=110)
ax.set_xlabel('Percentage of population exposure', fontsize=110)

# Add space between the bars
ax.set_ylim(-bar_height, len(df_pop["population exposure"]) - bar_height)
ax.spines[['right', 'top']].set_visible(False)

# Add the percentage values on the bars in black
for bar in bars:
    value = bar.get_width()
    ax.text(value, bar.get_y() + bar.get_height() / 2, f'{value:.1f}%', ha='left', va='center', color='black', fontsize=100)

plt.tight_layout()
plt.savefig("Population_exposure.png")
```
